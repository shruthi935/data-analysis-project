```python
import pandas as pd

file_path = '/content/Hotel Booking Demand Dataset..,,,,,,,.csv'

# Step 1 â€” Try reading with flexible settings
data = pd.read_csv(
    file_path,
    sep=',',                # Assume CSV is comma-separated
    quotechar='"',          # Handle text with commas
    on_bad_lines='skip',    # Skip badly formatted rows
    engine='python'         # Python engine handles irregular rows better
)

# Step 2 â€” Remove leading/trailing spaces in column names
data.columns = data.columns.str.strip()

# Step 3 â€” If 'is_canceled' not in columns, check the headers
if 'is_canceled' not in data.columns:
    print("Warning: 'is_canceled' column not found, here's the actual header list:")
    print(data.columns.tolist())

# Step 4 â€” Preview loaded data
print("Shape of dataset:", data.shape)
print("Sample rows:")
print(data.head())
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=data, x='is_canceled')
plt.title('Booking Cancellation Count')
plt.xlabel('Is Canceled (1 = Yes, 0 = No)')
plt.ylabel('Number of Bookings')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=data, x='hotel')
plt.title('Bookings by Hotel Type')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(
    data=data,
    x='arrival_date_month',
    order=['January','February','March','April','May','June',
           'July','August','September','October','November','December']
)
plt.xticks(rotation=45)
plt.title('Monthly Booking Trends')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=data, x='hotel', y='adr')
plt.title('Average Daily Rate by Hotel Type')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=data, x='market_segment', hue='is_canceled')
plt.xticks(rotation=45)
plt.title('Market Segment vs Cancellation')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data['lead_time'], bins=50, kde=True)
plt.title('Lead Time Distribution')
plt.xlabel('Days Between Booking & Arrival')
plt.show()
```

```python
if 'meal' in data.columns:
    meal_counts = data['meal'].value_counts()
    plt.figure(figsize=(6,6))
    plt.pie(meal_counts, labels=meal_counts.index, autopct='%1.1f%%', startangle=90)
    plt.title('Meal Type Distribution')
    plt.show()
```

```python
if {'hotel','is_canceled'}.issubset(data.columns):
    cross_tab = pd.crosstab(data['hotel'], data['is_canceled'])
    cross_tab.plot(kind='bar', stacked=True, figsize=(8,5), colormap='Paired')
    plt.title('Hotel Type vs Cancellation')
    plt.xlabel('Hotel Type')
    plt.ylabel('Number of Bookings')
    plt.legend(title='Is Canceled', labels=['No', 'Yes'])
    plt.show()
```

```python
if {'adr','lead_time'}.issubset(data.columns):
    plt.figure(figsize=(8,5))
    sns.scatterplot(data=data, x='lead_time', y='adr', alpha=0.5)
    plt.title('ADR vs Lead Time')
    plt.xlabel('Lead Time (days)')
    plt.ylabel('Average Daily Rate')
    plt.show()
```

```python
if {'customer_type','adr'}.issubset(data.columns):
    plt.figure(figsize=(8,5))
    sns.boxplot(data=data, x='customer_type', y='adr')
    plt.title('ADR by Customer Type')
    plt.show()
```
